```python
import matplotlib
```

```python
speedup_regular = [1, 1.78, 2.504, 3.35]
speedup_disable = [1, 2.03, 3.16, 5.9]
threads = [1, 2, 4, 8]

import matplotlib.pyplot as plt

plt.scatter(threads, speedup_regular, color='blue', label='Regular')
plt.scatter(threads, speedup_disable, color='red', label='Disable')
plt.xlabel('Threads')
plt.ylabel('Speedup')
plt.title('Speedup vs Threads')
plt.legend()
plt.show()
```
